 To determine whether an employee gets a bonus based on the conditions:

Sales > 40,000

Attendance  90%

you can use the IF function combined with AND.

 Excel Formula

Assuming:

Sales is in cell B2

Attendance is in cell C2

    =IF(AND(B2>40000,C2>=90),"Bonus","No Bonus")

 . Explanation:

AND(B2>40000, C2>=90) checks if both conditions are met.

If both are true  returns "Bonus"

If either condition is false  returns "No Bonus"

Example Results:

Arjun  Bonus

Simran  No Bonus

Rohit  No Bonus

Anjali  No Bonus

Marks are in cell B2.

     Nested IF Formula (Excel)
=IF(B2>=90,"A",IF(B2>=75,"B",IF(B2>=50,"C","Fail")))

 How the Formula Works (Logic)

Excel checks the conditions from left to right:

IF(B2>=90,"A",  )
 If marks are 90 or more, result is A

If not, Excel moves to:
IF(B2>=75,"B",  )
 If marks are 75 or more, result is B

If not, Excel moves to:
IF(B2>=50,"C","Fail")
 If marks are 50 or more, result is C
 Otherwise, result is Fail

  Correct Answer

Department of Employee ID E103 = Finance

 Which LOOKUP Function is Used?
VLOOKUP is used.
 Formula (Excel)

Assuming:

EmpID is in column A

Department is in column C

Table range is A2:C5

    =VLOOKUP("E103",A2:C5,3,FALSE)

 Explanation of the Formula

"E103"  Value to search

A2:C5  Table range

3  Column number of Department

FALSE  Exact match required

Excel searches for E103 in the first column (EmpID) and returns the value from the Department column.

 Why VLOOKUP is Used?

 EmpID is in the first (leftmost) column
 We are retrieving data from right to left
 Exact match is needed

 ** Answer** -


  Which Commission Rate Applies?

45,000 falls between 20,000 and 50,000,
so the applicable commission rate = 10%.

 Why VLOOKUP is Used?

The commission table is range-based

Sales Amount column is in ascending order

We need the nearest lower value, not an exact match

 Therefore, VLOOKUP with approximate match (TRUE) is used.

 VLOOKUP Formula (Excel)

Assuming:

Sales amount (45,000) is in cell E2

Commission table range is A2:B5

    =VLOOKUP(E2,A2:B5,2,TRUE)

 How the Formula Works

Excel looks for 45,000 in the first column

Since exact match is not found, it selects the largest value less than 45,000, which is 20,000

It returns the corresponding commission rate  10%

    INDEX(B2:B5, 2)

    XLOOKUP Formula (Excel)

Assuming:

EmpID is in A2:A5

Department is in B2:B5

    =XLOOKUP("E303", A2:A5, B2:B5)

 How the Formula Works

"E303"  Value to look up

A2:A5  Lookup array (EmpID column)

B2:B5  Return array (Department column)

XLOOKUP searches for E303 in the EmpID column and returns the corresponding value from the Department column.
