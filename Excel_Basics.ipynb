**1.Relative Cell Referencing:** A relative reference changes automatically when a
formula is copied to another cell.

For Example: Assume the data is entered in Excel as follows:

Cell A2 contains Car

Cell B2 contains 500000

Cell A3 contains Bicycle

Cell B3 contains 10500

Cell A4 contains Auto-Rickshaw

Cell B4 contains 25000






Now, you want to calculate 10% tax on the price.

In cell C2, you enter the formula:

    =B2*10%


When you copy this formula downward:

In cell C3, the formula automatically changes to:

    =B3*10%


In cell C4, the formula automatically changes to:

    =B4*10%


This happens because relative cell referencing adjusts the cell reference based on the new row.



**2.Absolute Cell Referencing:**  An absolute reference does not change when the formula is copied. It uses $ to lock the row and column.


For Example: Suppose the worksheet contains a fixed tax rate:

Cell D1 contains 10%

The price data is entered as follows:

Cell B2 contains 500000

Cell B3 contains 10500

Cell B4 contains 25000

Now, to calculate tax using the fixed tax rate:

In cell C2, you enter the formula:

    =B2*$D$1


When you copy this formula downward:

In cell C3, the formula becomes:

    =B3*$D$1


In cell C4, the formula becomes:

    =B4*$D$1


Even though the price cell changes from B2 to B3 to B4, the reference $D$1 remains the same because it is an absolute cell reference.


**3.Mixed Cell Referencing:** Locks only the row or only the column, not both.

Combination of relative and absolute.

● $A1 → column A fixed, row changes.

● A$1 → row 1 fixed, column changes.

If tax rates are placed in row 1 and prices are in column B:

    =B2*D$1

Row 1 stays fixed

Column changes if copied sideways

Useful when working with tables and multiple formulas.

**Answer =** Given data (assume it is arranged like this):

A2: Item A, B2: 90, C2: 1000

A3: Item B, B3: 150, C3: 1200

A4: Item C, B4: 250, C4: 1500

Here:

Column B = Price

Column C = Sales


**● Condition:**

We need to calculate the average sales for items that meet both conditions:

Price is greater than 100

Price is less than 300


**● Identify eligible items:**

Item A → Price = 90 ❌ (not above 100)

Item B → Price = 150 ✅

Item C → Price = 250 ✅

So, we calculate the average of sales for Item B and Item C.

Excel Formula:

    =AVERAGEIFS(C2:C4, B2:B4, ">100", B2:B4, "<300")


**● Explanation of the formula:**

C2:C4 → Reminder: Sales range (what we want to average)

B2:B4 → Price range (used for conditions)

">100" → First condition (price greater than 100)

"<300" → Second condition (price less than 300)

The formula calculates the average sales of items priced between 100 and 300, which in this case are Item B and Item C.
